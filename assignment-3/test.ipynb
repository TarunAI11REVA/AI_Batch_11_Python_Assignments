```python
import numpy as np
import pandas as pd

cars = pd.read_csv('mtcars.csv')
cars.head()
cars.tail()
cars.describe()

type(cars)

cars.shape # outputs the dimensions of the DataFrame (rows, columns) in out case it will be (32, 12)

cars.columns # outputs the column names of the DataFrame

cars.dtypes # outputs the data types of each column

cars.T # for transposing the DataFrame (swapping rows and columns)

sorted_cars = cars.sort_index(axis=1, ascending=False) # sorts the DataFrame by its index in descending order
sorted_cars.head()

cars_sorted_by_mpg = cars.sort_values(by='mpg', ascending=False) # sorts the DataFrame by the 'mpg' column in descending order
cars_sorted_by_mpg.head()

cars_sorted_by_mpg_ascending = cars.sort_values(by='mpg', ascending=True) # sorts the DataFrame by the 'mpg' column in ascending order
cars_sorted_by_mpg.head()

cars['displ/cyl'] = cars['disp'] / cars['cyl'] # creates a new column 'displ/cyl' by dividing 'disp' by 'cyl'
cars['price'] = np.nan # adds a new column 'price' with NaN values
cars.head() 

cars.loc[:, ['mpg', 'cyl', 'disp']] # selects all rows and specific columns 'mpg', 'cyl', and 'disp'

cars.loc[0:4, ['mpg', 'cyl', 'disp']] # selects the first 5 rows and specific columns 'mpg', 'cyl', and 'disp'

cars.iloc[0:5, 0:3] # selects the first 5 rows and the first 3 columns using integer-location based indexing

cars.iloc[0:5, : ] # selects the first 5 rows and all columns using integer-location based indexing

cars.iloc[0:5, [0,2,9]] # selects the first 5 rows and specific columns (0, 2, and 9) using integer-location based indexing

cars.iloc[16, 1] # selects the row with index 18 and all columns using label-based indexing


```
