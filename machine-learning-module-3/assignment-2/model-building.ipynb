```python
# MODEL EVALUATION (Classification Models)
# Evaluate models using metrics discussed in class:
# - Accuracy
# - Confusion Matrix
# - Precision, Recall, F1-score
# - ROC-AUC + ROC Curve

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    roc_curve
)

# Create a reusable evaluation function
# This function:
# - Takes a trained model + test data
# - Generates predictions
# - Calculates all key metrics
# - Plots confusion matrix
# - Returns metrics in dictionary form for comparison table
def evaluate_classifier(model, X_test, y_test, model_name="Model"):
    """
    Evaluates a classification model using common metrics and plots.
    Returns a dictionary of key metrics for summary comparison.
    """
    
    # Predict class labels (0/1)
    y_pred = model.predict(X_test)
    
    # Predict probabilities for positive class (Survived = 1)
    # Some models provide predict_proba. If available, we use it for ROC-AUC.
    # If not available, ROC-AUC cannot be computed in probability form.
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
    else:
        y_prob = None
    
    # Compute core metrics
    acc = accuracy_score(y_test, y_pred)                 # Overall correctness
    prec = precision_score(y_test, y_pred, zero_division=0)  # TP/(TP+FP)
    rec = recall_score(y_test, y_pred, zero_division=0)      # TP/(TP+FN)
    f1 = f1_score(y_test, y_pred, zero_division=0)           # Harmonic mean of precision & recall
    
    # ROC-AUC requires probabilities (best practice)
    auc = roc_auc_score(y_test, y_prob) if y_prob is not None else np.nan
    
    # Print results
    print(f"\n==================== {model_name} ====================")
    print(f"Accuracy  : {acc:.4f}")
    print(f"Precision : {prec:.4f}")
    print(f"Recall    : {rec:.4f}")
    print(f"F1-score  : {f1:.4f}")
    print(f"ROC-AUC   : {auc:.4f}" if y_prob is not None else "ROC-AUC   : Not available (no predict_proba)")
    
    # Confusion Matrix
    # Confusion matrix format:
    # [[TN, FP],
    #  [FN, TP]]
    # TN: True Negatives, FP: False Positives
    # FN: False Negatives, TP: True Positives
    # -------------------------
    cm = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix:\n", cm)
    
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.show()
    
    # Classification Report
    # Provides per-class precision, recall, f1-score and support
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, zero_division=0))
    
    # Return key metrics for summary table
    return {
        "Model": model_name,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1-score": f1,
        "ROC-AUC": auc
    }


# Evaluate all trained models
# Make sure these model variables exist in your notebook.
# If your names are different, replace them accordingly.
metrics_list = []

# Logistic Regression (tuned or baseline)
# Replace best_lr with your model name if needed
metrics_list.append(evaluate_classifier(best_lr, X_test, y_test, "Logistic Regression"))

# Decision Tree Classifier (tuned or pruned)
metrics_list.append(evaluate_classifier(best_dt, X_test, y_test, "Decision Tree Classifier"))

# Create a comparison table for all models
# This table helps you clearly justify which model performs best.
results_df = pd.DataFrame(metrics_list)

# Sort by ROC-AUC (common and reliable metric for classification ranking)
results_df_sorted = results_df.sort_values(by="ROC-AUC", ascending=False)

print("\n==================== MODEL COMPARISON TABLE ====================")
display(results_df_sorted)


# Plot ROC Curves for all models on one graph
# ROC curve compares True Positive Rate vs False Positive Rate.
# AUC summarizes the ROC curve (higher is better).
plt.figure(figsize=(7,6))

for model, name in [(best_lr, "Logistic Regression"),
                    (best_dt, "Decision Tree")]:
    
    # Only plot ROC if model supports predict_proba
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        auc = roc_auc_score(y_test, y_prob)
        plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.2f})")

# Random classifier diagonal reference
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.show()

# - Accuracy: overall correctness, but can be misleading if classes are imbalanced
# - Precision: reliability of predicted survivors
# - Recall: ability to catch actual survivors
# - F1-score: balance between precision and recall
# - ROC-AUC: best overall metric to compare ranking/class separation performance
```
