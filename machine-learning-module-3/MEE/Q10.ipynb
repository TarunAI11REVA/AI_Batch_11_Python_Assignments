# Q10

Define the following parameters/components of a time series model and explain what they signify:

a) Trend  
b) Seasonality  
c) Cyclicality  
d) Random Noise


## a) Trend
**Definition:** The long-term direction or movement in the time series over an extended period.

**Significance:**
- Shows whether the series is generally increasing, decreasing, or staying stable over time.
- Often reflects underlying growth/decline drivers (e.g., inflation, population growth, technology adoption).
- Helps in forecasting long-run behavior.


## b) Seasonality
**Definition:** Regular and repeating patterns that occur at fixed and known intervals (e.g., daily, weekly, monthly, quarterly, yearly).

**Significance:**
- Captures predictable fluctuations due to the calendar, weather, holidays, business cycles, etc.
- Example: higher ice cream sales every summer; increased retail sales during festivals/holidays.
- Important for selecting seasonal models (e.g., SARIMA) and for seasonal adjustment.


## c) Cyclicality
**Definition:** Repeating rises and falls in the series that occur over longer, non-fixed periods (the cycle length is not strictly regular).

**Significance:**
- Often linked to broader economic or business cycles (booms and recessions).
- Unlike seasonality, cycles do **not** have a constant frequency and can vary in duration and intensity.
- Helps explain medium-to-long term fluctuations beyond seasonal effects.


## d) Random Noise (Irregular Component)
**Definition:** Unpredictable, residual variation in the time series after removing trend, seasonality, and cyclical effects.

**Significance:**
- Represents random shocks, measurement errors, and unforeseen events.
- Ideally behaves like “white noise” (mean ~ 0, no autocorrelation).
- High noise makes forecasting harder; analyzing noise helps check model adequacy.
